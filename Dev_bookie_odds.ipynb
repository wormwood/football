```python
def swap (row):
    if row[0]!= row[2]:
        return[row[0], row[1],row[2],row['AwayOdds'], row['DrawOdds'], row['HomeOdds']]
    else:
        return[row[0], row[1],row[2],row['HomeOdds'], row['DrawOdds'], row['AwayOdds']]
```
